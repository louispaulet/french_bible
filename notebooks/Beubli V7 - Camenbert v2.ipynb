# Sentence Masking functions

We need some functions to handle the following tag structure : `<mask:il>`, where `<mask>` marks the place where CamenBERT has to make a guess and `il` is an example of PRO:PERS. 

In our case, CamenBERT only needs to guess that `<mask:il>` cases, other masks (ie. `<mask:elle>` or ``<mask:ils>`) are not needed for guessing. However, we still add them to the text as they will be easy to match when doing the final replacement.

```python
def add_masks(tagTree):

  final_sentence = []
  for tag in tagTree:
    # print(tag.what)
    if (isinstance(tag, treetaggerwrapper.NotTag)):
      # print(tag.what)
      if (tag.what == '<ttpw:space />'):
        # print(' ')
        final_sentence.append(' ')
    if (isinstance(tag, treetaggerwrapper.Tag)):
      # print(tag.word, tag.pos)
      if ((tag.pos == 'PRO:PER') and (tag.word.lower() in ['il', 'elle', 'ils', 'elles'])):
        final_sentence.append('<mask:'+tag.word+'>')

      else:
        final_sentence.append(tag.word)
      

  return ''.join(final_sentence)

final_sentence = add_masks(tagList[14671])
final_sentence
```

This function counts the number of masks and can filter on input PRO:PER

```python
# we need to count the number of tags in a sentece, with a filter on PRO:PER
def count_masks(input_sentence, input_pro_per):
  return input_sentence.count('mask:'+input_pro_per)

# show total count of masks, as welle as count with critical case "il"
count_masks(final_sentence, ''), count_masks(final_sentence, 'il')
```

```python
final_sentence
```

```python
def get_single_mask_at(input_sentence, mask_position):

  exploded_list = re.compile(r'(\<\bmask\:(?:\bil|\bIl|\belle|\bElle|\bils|\bIls|\belles|\bElles)\>)').split(input_sentence)
  final_sentence = []
  il_pro_per_counter = 0

  for sentence_piece in exploded_list:
    if ('mask:' in sentence_piece):
      pro_per = re.compile(r'\<\bmask:(.*)\>').findall(sentence_piece)
      pro_per = pro_per[0]
      # case of il pro:per
      if (pro_per.lower() == 'il'):
        if (il_pro_per_counter == mask_position):
          final_sentence.append('<mask>')
        else:
          final_sentence.append(pro_per)

        il_pro_per_counter = il_pro_per_counter+1
      else:
        # case of other pro:per
        final_sentence.append(pro_per)
    else:
      final_sentence.append(sentence_piece)
    

  return ''.join(final_sentence)


print(get_single_mask_at(final_sentence, 0))
print(get_single_mask_at(final_sentence, 1))
```

Then we can combine both functions to get all permutations of "il" masks:

/!\ Case sensitive counting function

```python
def get_all_problematic_permutations(input_sentence):

  total_problematic_mask_count = count_masks(input_sentence, 'il') + count_masks(input_sentence, 'Il')

  total_permutation_list = []
  for cpt in range(0, total_problematic_mask_count):
    total_permutation_list.append(get_single_mask_at(input_sentence, cpt))

  return total_permutation_list

get_all_problematic_permutations(final_sentence)
```

# CamenBERT guesser

We will guess a single mask per sentence, then guess for all sentence combinations of masked words.  

Before guessing if PRO:PER is replaceable or not, we will calibrate with a very small curated dataset.

## Unitary guesser

```python
def guess_single_mask_in_sentence(input_sentence):
  
  
  camenbert_results = camembert.fill_mask(input_sentence, topk=5)
  camenbert_predicted_tags = []
  for camenbert_result in camenbert_results:
    camenbert_predicted_tags.append(camenbert_result[2])
  return set(camenbert_predicted_tags)


for test_test_test_sentence in get_all_problematic_permutations(final_sentence):
  print(guess_single_mask_in_sentence(test_test_test_sentence))
```

We can see in the example above that the proposed words are all subjects. In both instances, the replacement is OK. 

## Small handcrafted dataset

```python
neutral_il_sentence_list = [
  "il pleut en ce moment",
  "il fait beau aujourd'hui",
  "il y eut un temps où les animaux",
  "il fut un temps où les animaux",
  "il m'arrive d'eternuer",
  "il m'est encore arrivé un truc",
]


gendered_il_sentence_list = [
  "il s'est levé du pied gauche",
  "Il est parti tôt ce matin",
  "elle est allée en bas",
  "elles sont en haut",
  "il s'est habillé",
  "ils sont en vacances",
]


def get_tag_set_for_sentence_list(sentence_list):
  for current_sentence in sentence_list:
    final_set = set()
    current_sentence = re.sub(r'(((Il|il|elle)s?)\s)', '<mask>', current_sentence)
    final_set = set(guess_single_mask_in_sentence(current_sentence)).union(final_set)
  return final_set


print(get_tag_set_for_sentence_list(neutral_il_sentence_list))

print(get_tag_set_for_sentence_list(gendered_il_sentence_list))
```

We can see the the neutral dataset yields neutral suggestions, while the gendered dataset yields gendered suggestions. There is a way to bootstrap a recognition system.

```python
result_list = []
for tag in tqdm(tagList_sample_200):
  masked_sentence = add_masks(tag)
  for problematic_sentence in get_all_problematic_permutations(masked_sentence):
    # print(problematic_sentence)
    if (problematic_sentence.count('mask')):
      # print('inside')
      tag_set = guess_single_mask_in_sentence(problematic_sentence)
      result_list.append([problematic_sentence, tag_set])

result_df = pd.DataFrame(result_list, columns=['mask_sentence', 'bert_tags'])
result_df
```

# Silhouette analysis

We want to have a fine tuned Kmeans model. To do this, we will find the ideal K number for a TF-IDF dataset (un-reduced).

```python
# import neccessaries librariesimport pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples,silhouette_score
from sklearn.decomposition import PCA


def silhouette_analysis(input_df):
  # # load petal data
  # data = datasets.load_iris()dir(data)
  # # load into Dataframe 
  # df = pd.DataFrame(data.data,columns = data.feature_names)
  # print(df.shape)
  # df.head()
  # input_df = df.drop(['sepal length (cm)', 'sepal width (cm)'],axis = 'columns')
  input_df.head()

  # perform radical PCA to get only 2 dimensions to plot ?
  # plot scatter plot
  # pca = PCA(n_components=2)
  # df_to_plot = pca.fit_transform(input_df)

  df_to_plot = TSNE(n_components=2, learning_rate='auto',
                   init='random').fit_transform(input_df.to_numpy())
  df_to_plot = pd.DataFrame(df_to_plot, columns=['dim1', 'dim2'])

  plt.scatter(df_to_plot['dim1'],df_to_plot['dim2'])

  # Now check silhouette coefficient
  for i,k in enumerate([2,3, 6, 12, 18, 24]):
      
      fig, ax = plt.subplots(1,2,figsize=(15,5))
      
      # Run the kmeans algorithm
      km = KMeans(n_clusters=k)
      y_predict = km.fit_predict(input_df)
      centroids  = km.cluster_centers_
      # get silhouette
      silhouette_vals = silhouette_samples(input_df,y_predict)
          #silhouette_vals
      # silhouette plot
      y_ticks = []
      y_lower = y_upper = 0
      for i,cluster in enumerate(np.unique(y_predict)):
        cluster_silhouette_vals = silhouette_vals[y_predict ==cluster]
        cluster_silhouette_vals.sort()
        y_upper += len(cluster_silhouette_vals)
        
        ax[0].barh(range(y_lower,y_upper),
        cluster_silhouette_vals,height =1);
        ax[0].text(-0.03,(y_lower+y_upper)/2,str(i+1))
        y_lower += len(cluster_silhouette_vals)
            
        # Get the average silhouette score 
        avg_score = np.mean(silhouette_vals)
        ax[0].axvline(avg_score,linestyle ='--',
        linewidth =2,color = 'green')
        ax[0].set_yticks([])
        ax[0].set_xlim([-0.1, 1])
        ax[0].set_xlabel('Silhouette coefficient values')
        ax[0].set_ylabel('Cluster labels')
        ax[0].set_title('Silhouette plot for the various clusters');
          
          
      # scatter plot of data colored with labels

      ax[1].scatter(df_to_plot['dim1'],
      df_to_plot['dim2'] , c = y_predict);
      ax[1].scatter(centroids[:,0],centroids[:,1],
      marker = '*' , c= 'r',s =250);
      ax[1].set_xlabel('Eruption time in mins')
      ax[1].set_ylabel('Waiting time to next eruption')
      ax[1].set_title('Visualization of clustered data', y=1.02)
      
      plt.tight_layout()
      plt.suptitle(f' Silhouette analysis using k = {k}',fontsize=16,fontweight = 'semibold')
      # plt.savefig(f'Silhouette_analysis_{k}.jpg')
      plt.show()

silhouette_analysis(tf_idf_matrix)
```
