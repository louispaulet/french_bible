## Testing neutral vs gendered sentences

```python
from pygrammalecte import grammalecte_text

texte_bidon = """\
Coucou, je veut du fromage.
Je sais coder en VHDL.
Le VHDL est est compliquer.
"""

for message in grammalecte_text(texte_bidon):
    print(message)
```

Gendered sentences throw an error when the pronoun is changed:

```python
texte_bidon = """\
Il est beau.
Elle est belle.
Elle est beau.
Il est belle.
"""

for message in grammalecte_text(texte_bidon):
    print(message)
```

Neutral sentences DO NOT throw an error when the pronoun is changed:

```python
texte_bidon = """\
Il semble que le temps se soit assombri.
Elle semble que le temps se soit assombri.
"""

for message in grammalecte_text(texte_bidon):
    print(message)
```

We can remplace PR:PERS tokens in sentences that throw `gv1__ppas_3sg_mas_verbe_état__b4_a1_1` errors
